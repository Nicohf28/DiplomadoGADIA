## 1. For
El bucle For se utiliza para iterar sobre una secuencia (como una lista, tupla, diccionario, conjunto o cadena de caracteres).
Se ejecuta una vez por cada elemento en la secuencia.

```python
# Imprimir todos los números de una secuencia de datos.

secuencia = [1, 2, 3, 4, 5]
for i in secuencia:
  print(i)
```

```python
# Clasificar los números que son pares y los que son impares, de la lista secuencia.
for i in secuencia:
  if i%2 == 0:
    print(f"// {i} es par.")
  else:
    print(f"// {i} es impar. ")
```

```python
# Suma
a = 10
b = 5
print(a + b)  # 15

# Resta
print(a - b)  # 5

# Multiplicación
print(a * b)  # 50

# División
print(a / b)  # 2.0

# División entera
print(a // b)  # 2

# Módulo
print(a % b)  # 0

# Exponenciación
print(a ** b)  # 100000

# Asignación compuesta
a += b
print(a)  # 15

# Redondeo
c = 3.14159
print(round(c, 2))  # 3.14

# Valor absoluto
d = -10
print(abs(d))  # 10

# Máximo y mínimo
e = [1, 2, 3, 4, 5]
print(max(e))  # 5
print(min(e))  # 1

# Potencia
print(pow(2, 3))  # 8
```

```python
# Supongamos que tienes una lista de precios de productos y quieres aplicar un descuento del 10% a cada precio.

precios = [100, 150, 200, 250, 300]
preciosConDescuento = []

for i in precios:
  preciosConDescuento.append(i * 0.9)

print(preciosConDescuento)
```

## 2. While
El bucle while se utiliza para repetir un bloque de código mientras una condición sea verdadera.

```python
# Si Agamoto destruye el planeta, Strange viene a negociar. Mientras Agamoto no destruya la tierra, estrange no viene a negociar.

#while condición:
  # Código a Ejecutar.
```

```python
expresion_bool = True

while expresion_bool:
  pregunta = input("-> ¿Quieres destruir el planeta?")

  if pregunta == "Si":
    print("// He venido a negociar. ")
    break
  elif pregunta == "No":
    print("// Aún no vengo a negociar. ")
  else:
    print("Entrada inválida. Intente de nuevo. ")
```

```python
# Construir un contador que vaya hasta 5 y cuando la variable tenga este valor, deje de contar.

contador = 1
while contador <= 5 :
  print(contador)
  contador += 1
```

**Plabra Reservada break:** Sale del bucle antes de que se evalúe la condición o se termine de iterar sobre la secuencia.


**Palabra Reservada continue:** Salta el resto del código dentro del bucle y vuelve a la condición o a la siguiente iteración.


**Palabra Reservada else (en Whiles):** Se ejecuta una vez cuando el bucle termina, a menos que se haya interrumpido con break.

```python
# Supongamos que quieres solicitar a un usuario que ingrese una contraseña hasta que lo haga correctamente.
contrasenia_correcta = "segura123"
intentos = 3

print("// Nota: Tienes 3 intentos para ingresar la contraseña correcta.")
while intentos > 0:
  contrasenia = input("-> Ingrese la contraseña: ")

  if contrasenia == contrasenia_correcta:
    print("// La contraseña es correta. Acceso concedido. ")
    break
  else:
    print(f"// La contraseña es incorrecta. Acceso Denegado. Te quedan {intentos-1} intentos. ")

    intentos -= 1

    if intentos == 0:
      print("// Se te acabaron los intentos. Se acabó.")
```

```python
# Escribe un programa que permita al usuario adivinar un número secreto entre 1 y 10.
# El programa debe seguir pidiendo al usuario que adivine hasta que acierte.

"""
Define un número secreto.
Utiliza un bucle while para seguir solicitando adivinanzas al usuario hasta que acierte.
Si el usuario adivina el número, imprime un mensaje de felicitación y termina el programa.
"""

num_secreto = 8
while True:
  num_ingresado = int(input("-> Ingresa un número del 1 al 10: "))

  if num_ingresado<1 or num_ingresado>10 :

    print("// El número ingresado debe estar entre el 1 y el 10. Intente de nuevo. ")

  else:

    if num_ingresado != num_secreto:
      print("// El número ingresado es: Incorrecto! Intente de nuevo!!")
    else:
      print("// El número ingresado es: Correcto! Ganaste!!")
      break

# Fin del Juego.
```

```python
# Recorte.

numero_correcto = 6
intento = 0

while True:
  intento = int(input("-> Ingrese un número entre 1 y 10: "))
  if intento>=1 and intento<=10:
    if intento == numero_correcto:
      print(f"// Felicidades! {intento} era el número secreto. ")
      break
    else:
      print(f"// {intento} no es el número secreto... Intenta de nuevo. ")
  else:
    print("// El número debe estar entre 1 y 10. Intenta de nuevo. ")
```

```python
# Sin break.

numero_correcto = 6
intento = 0

while intento != numero_correcto:
  intento = int(input("-> Ingrese un número entre 1 y 10: "))
  if intento>=1 and intento<=10:
    if intento == numero_correcto:
      print(f"// Felicidades! {intento} era el número secreto. ")
      break
    else:
      print(f"// {intento} no es el número secreto... Intenta de nuevo. ")
  else:
    print("// El número debe estar entre 1 y 10. Intenta de nuevo. ")
```

```python
# Sin número repetidos:
# Sin break.

numero_correcto = 6
intento = 0
numeros_colocados = []

while intento != numero_correcto:
  intento = int(input("-> Ingrese un número entre 1 y 10: "))
  if intento>=1 and intento<=10:
    if intento not in numeros_colocados:
      if intento == numero_correcto:
        print(f"// Felicidades! {intento} era el número secreto. ")
        break
      else:
        print(f"// {intento} no es el número secreto... Intenta de nuevo. ")
        numeros_colocados.append(intento)
    else:
      print("// Este número ya ha sido utilizado. Intenta con otro. ")
      print("// Los números ya utilizados son: ", numeros_colocados)
  else:
    print("// El número debe estar entre 1 y 10. Intenta de nuevo. ")
```
